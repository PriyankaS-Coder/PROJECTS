# **Factors  Affecting BMI**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Diet.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
BMI = df['Weight in (kg)'] / (df['Height  in (m)'])**2
BMI
```

```python
df = df.assign(BMI = BMI)
df
```

```python
df.dropna().sum()
```

```python
def get_bmi_category(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return "Underweight"
    elif bmi > 24.9:
        return "Overweight"
    else:
        return "Normal"

df['BMI_Category'] = df['BMI'].apply(get_bmi_category)
df
```

```python
exercise_mapping = {
    "Never": 0,
    "Rarely": 1,
    "Thrice a week": 3,
    "Daily": 7
}
df['Weekly_Excercise_Numeric'] = df['Weekly Excercise'].map(exercise_mapping)
df
```

```python
junk_food = {
    "Never": 0,
    "Rarely": 1,
    "1-3 times a week": 3, # Corrected key
    "4-6 times a week": 6,
    "Daily": 7
}
df["Junk Food"] = df["Junk Food Frequency"].map(junk_food)
df
```

```python
Fruits_veg = {
    "Never": 0,
    "Rarely": 1,
    "1-3 times a week": 3, # Corrected key
    "4-6 times a week": 6,
    "Daily": 7
}
df["Fruits Veg"] = df["Dietary Intake of Fruits / Vegetables"].map(Fruits_veg)
df
```

```python
fig, ax = plt.subplots()
bmi = df.groupby("Weekly_Excercise_Numeric")["BMI"].sum()
ax.scatter(bmi.index,  bmi.values, marker = "o")
ax.set_xlabel("Weekly Excercise")
ax.set_ylabel("Total BMI")
ax.set_title("Total BMI by Weekly Excercise")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 5))
ax[0].bar(df["Junk Food"],  df["BMI"], color = "blue")
ax[0].set_xlabel("Junk Food (Numerical)")
ax[0].set_ylabel("BMI")
ax[0].set_title("BMI by Junk Food Frequency")

ax[1].bar(df["Fruits Veg"],  df["BMI"], color = "red")
ax[1].set_xlabel("Fruits Veg (Numerical)")
ax[1].set_ylabel("BMI")
ax[1].set_title("BMI by Fruits/Vegetable Intake")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

# Chi2 Test

Ho - There is a relationship between BMI category and Junk food

Ha - There is no relationship between BMI category and Junk Food

```python
data_table = pd.crosstab(df["Junk Food"], df["BMI_Category"])
data_table
```

```python
observed_values = data_table.values
observed_values
```

```python
from scipy.stats import chi2_contingency
expected_values = chi2_contingency(observed_values)
expected_values
```

```python
from scipy.stats import chi2
for i in zip(observed_values, expected_values):
  print(i)
```

```python
chi2_test = [(o-e)**2/e for o, e in zip(observed_values, expected_values)]
chi2_test
```

```python
alpha = 0.05
```

```python
dof = 3
```

```python
statistic = 1.276
```

```python
chi2_critical= chi2.ppf(1-alpha, dof)
chi2_critical
```

```python
if statistic >= chi2_critical :
  print("Reject Null Hypothesis")
else:
  print("Accept Null Hypothesis")
```
