```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

# A significant portion of customers have churned, indicating that customer retention is a major challenge. The churn rate is high enough to require strategic intervention.

```python
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

# There is no major difference in churn rate between males and females. Gender is not a strong predictor of churn.

```python
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title("Churn by Senior Citizen")
plt.show()
```

#Senior citizens have a much higher churn rate compared to non-senior customers. This indicates older customers may require better support or simpler plans.

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=20)
plt.show()
```

# Customers with Month-to-Month contracts churn the most, whereas those with 1-year or 2-year contracts rarely churn. Long-term contracts improve retention significantly.

```python
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

#  Customers using Electronic Check churn at the highest rate. This payment method may be associated with customers who are price-sensitive or face payment difficulties

```python
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.title("Tenure vs Monthly Charges (Colored by Churn)")
plt.show()
```

```python
sns.countplot(x='InternetService', data=df)
plt.title("Internet Service Types")
plt.show()
```

# Fiber optic customers often show higher churn compared to DSL or those without internet service. This might be due to higher pricing or service dissatisfaction.

```python
sns.countplot(x='OnlineSecurity', hue='Churn', data=df)
plt.title("Churn by Online Security")
plt.show()
```

```python
sns.countplot(x='TechSupport', hue='Churn', data=df)
plt.title("Churn by Tech Support")
plt.show()
```

```python
sns.countplot(x='StreamingTV', hue='Churn', data=df)
plt.title("Churn by Streaming TV")
plt.show()
```

# Customers without Online Security churn more Security add-ons improve retention.Lack of Tech Support strongly correlates with higher churn.Support quality is crucial for retaining customers.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
plt.figure(figsize=(6,4))
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['MonthlyCharges'], bins=30, kde=True)
plt.title("Monthly Charges Distribution")
plt.show()
```

```python
sns.histplot(df['TotalCharges'], bins=30, kde=True)
plt.title("Total Charges Distribution")
plt.show()
```
