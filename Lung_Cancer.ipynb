# CHI2 TEST

Ho - There is a relationship between smoking and lung cancer

Ha - There is no relationship between smoking and lung cancer


```python
cross_table = pd.crosstab(df['SMOKING'], df['LUNG_CANCER'])
cross_table
```

```python
observed_values = cross_table.values
observed_values
```

```python
import scipy.stats as stats
expected_values = stats.chi2_contingency(observed_values)
expected_values
```

```python
from scipy.stats import chi2
for i in zip(observed_values,expected_values):
  print(i)
```

```python
chi2_test = [(o-e)**2/e for o, e in zip(observed_values, expected_values)]
chi2_test
```

```python
alpha = 0.05
```

```python
dof = 1
```

```python
statistic=np.float64(0.7225127713920805)
```

```python
chi2_critical= chi2.ppf(1-alpha, dof)
chi2_critical
```

```python
if statistic >= chi2_critical :
  print("Reject Null Hypothesis")
else:
  print("Accept Null Hypothesis")
```
